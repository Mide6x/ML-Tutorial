```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
data = pd.read_csv('bhp.csv')
```

```python
plt.hist(data['price_per_sqft'], bins=20, color='blue', edgecolor='black', rwidth=0.8)

# Adding labels and title
plt.xlabel('Price per Sqft (Naira)')
plt.ylabel('Count')
plt.yscale('log')
plt.title('Data')

# Show the plot
plt.show()
```

1. Applying Percentile

```python
lower_limit, upper_limit = data.price_per_sqft.quantile([0.001, 0.999])
lower_limit, upper_limit
```

```python
data[(data['price_per_sqft']>upper_limit)]
```

```python
data[(data['price_per_sqft']<lower_limit)]
```

```python
no_outlier = data[(data['price_per_sqft']>upper_limit) | (data['price_per_sqft']<lower_limit)]

no_outlier
```

```python
no_outlier.shape
```

```python
data2 = data[(data['price_per_sqft']<upper_limit) & (data['price_per_sqft']>lower_limit)]
```

```python
data.shape[0] - data2.shape[0]
```

2. Using z score

```python
data2['z_score'] = (data2['price_per_sqft'] - data2['price_per_sqft'].mean())/data2['price_per_sqft'].std()
```

```python
data2
```

```python
outlierss = data2[(data2.z_score>4) | (data2.z_score<-4)]

outlierss.shape
```

3. Remove using 4 Standard deviation

```python
price_data = data2['price_per_sqft']

# Calculate mean and standard deviation
mu = price_data.mean()
sigma = price_data.std()
```

```python
max_limit = mu + 4*sigma
min_limit = mu - 4*sigma
```

```python
max_limit, min_limit
```

```python
no_outlier_std = data2[(data2['price_per_sqft']>max_limit) | (data2['price_per_sqft']<min_limit)]

no_outlier_std.shape
```

```python
data3 = data2[(data2['price_per_sqft']<max_limit) & (data2['price_per_sqft']>min_limit)]

data3.shape
```
